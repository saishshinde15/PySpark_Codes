```python
from pyspark.sql import SparkSession
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("EDA").getOrCreate()
```

```python
df=spark.read.csv('/content/Facebook Metrics of Cosmetic Brand.csv',header=True,inferSchema=True)
```

```python
df.show()
```

```python
from pyspark.sql.functions import isnan, count, when, col #to calculte the null values in the dataframe
df.select([count(when(col(c).isNull(), c)).alias(c) for c in df.columns]).show()
```

```python
from pyspark.sql.functions import rand, when     #create some null values in the table

# Create a new DataFrame with null values in some columns
df = df.withColumn("Total Interactions", when(rand() < 0.1, None).otherwise(col("Total Interactions"))) \
                  .withColumn("Lifetime Post Total Reach", when(rand() < 0.05, None).otherwise(col("Lifetime Post Total Reach")))

# Display some rows to verify
df.show()

# Recalculate null counts
df.select([count(when(col(c).isNull(), c)).alias(c) for c in df.columns]).show()
```

```python
df=df.drop('_c0') # drop column
```

```python
df.show()
```

```python
df.na.drop().show() # drop null values
```

```python
df.na.drop(how='all').show() ## drops the entire row or column if all the data in that particular row or columns are null
```

```python
df.na.drop(how='any').show() # drops the null values (default settings)
```

```python
df.na.drop(how='any',thresh=2).show() # will drop the row in which there are atleast two non null values
```

```python
df.na.drop(how='any',subset=['Lifetime Post Total Reach']).show() # deleting nan values in the a particular column
```
